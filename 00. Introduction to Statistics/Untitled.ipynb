```python
def grouped_frequency_distribution(data, num_classes):
    data_range = max(data) - min(data)
    class_width = data_range / num_classes
    lower_limits = [min(data) + i * class_width for i in range(num_classes)]
    
    frequency_distribution = {f"{int(lower_limits[i])} - {int(lower_limits[i] + class_width)}": 0 for i in range(num_classes)}

    for value in data:
        for i in range(num_classes):
            if value >= lower_limits[i] and value < lower_limits[i] + class_width:
                frequency_distribution[f"{int(lower_limits[i])} - {int(lower_limits[i] + class_width)}"] += 1

    print("Grouped Frequency Distribution:")
    for interval, frequency in frequency_distribution.items():
        print(f"{interval}: {frequency}")
```

```python
def group_frequency(data):
    # Step 1: Find the Range
    data_range = np.max(data) - np.min(data)

    # Step 2: Calculate Class Width
    num_classes = 7
    class_width = data_range / num_classes

    # Step 3: Determine Class Limits
    class_limits = [np.min(data) + class_width * i for i in range(num_classes + 1)]

    # Step 4: Tally the Frequencies
    frequencies = np.histogram(data, bins=class_limits)[0]

    # Step 5: Display the Grouped Frequency Distribution
    print("Class Interval    Frequency")
    print("-" * 30)
    for i in range(num_classes):
        lower_limit = class_limits[i]
        upper_limit = class_limits[i + 1]
        print(f"{lower_limit:.2f} - {upper_limit:.2f}      {frequencies[i]}")
    print("-" * 30)
```
