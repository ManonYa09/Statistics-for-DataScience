```python
import numpy as np
y = ['A', 'B', 'B', 'AB', 'O', 'O', 'O', 'B',
     'AB', 'B', 'B', 'B', 'O', 'A', 'O', 'A', 
     'O', 'O', 'O', 'AB', 'AB', 'A', 'O', 'B', 'A']
unique_values, counts= np.unique(y, return_counts=True)
total_counts = np.sum(counts)

for value, count in zip(unique_values, counts):
    relative_frequency = count / total_counts
    relative_frequency = round(relative_frequency, 2)
    print(f'Value:{value}, Count:{count}, Relative frequency: {relative_frequency}')
```

```python
temparatures = [112, 100, 127, 120, 134, 118, 105, 110, 
               109, 112, 110, 118, 117, 116, 118, 122,
               114, 114, 105, 109, 107, 112, 114, 115, 
               118, 117, 118, 122, 106, 110, 116, 108, 
               110, 121, 113, 120, 119, 111, 104, 111,
               120, 113, 120, 117, 105, 110, 118, 112,
               114, 114]

a, b= np.histogram(temparatures, bins=7)
print(f'Frequency:{a} class{b}')
```

```python
temparatures = [112, 100, 127, 120, 134, 118, 105, 110, 
               109, 112, 110, 118, 117, 116, 118, 122,
               114, 114, 105, 109, 107, 112, 114, 115, 
               118, 117, 118, 122, 106, 110, 116, 108, 
               110, 121, 113, 120, 119, 111, 104, 111,
               120, 113, 120, 117, 105, 110, 118, 112,
               114, 114]
plt.hist(temparatures, bins=7)
```

```python

# Calculate the histogram and class intervals
frequencies, class_intervals = np.histogram(temperatures, bins=7)
class_intervals = [(class_intervals[i], class_intervals[i + 1]) for i in range(len(class_intervals) - 1)]

# Display the grouped frequency distribution
print("Class Interval      Frequency")
print("-" * 32)
for interval, frequency in zip(class_intervals, frequencies):
    lower_limit, upper_limit = interval
    print(f"{lower_limit:.2f} - {upper_limit:.2f}        {frequency}")
print("-" * 32)
```

```python

def group_frequency(data):
    # Step 1: Find the Range
    data_range = np.max(data) - np.min(data)

    # Step 2: Calculate Class Width
    num_classes = 7
    class_width = data_range / num_classes

    # Step 3: Determine Class Limits
    class_limits = [np.min(data) + class_width * i for i in range(num_classes + 1)]

    # Step 4: Tally the Frequencies
    frequencies = np.histogram(data, bins=class_limits)[0]

    # Step 5: Display the Grouped Frequency Distribution
    print("Class Interval    Frequency")
    print("-" * 30)
    for i in range(num_classes):
        lower_limit = class_limits[i]
        upper_limit = class_limits[i + 1]
        print(f"{lower_limit:.2f} - {upper_limit:.2f}      {frequencies[i]}")
    print("-" * 30)
```
